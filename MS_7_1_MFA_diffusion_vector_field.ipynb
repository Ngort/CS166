```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

```python
#for i in range(1,5):
alpha = 0.75
beta = 0.9
gamma = 5

diffop, bond = np.meshgrid(np.arange(-1, 1, .1), np.arange(0, 1, .1))
delta_o = alpha*-1*diffop*bond
delta_w = beta*bond*(1-bond)*(1-gamma*abs(diffop))

plt.figure()
plt.title(r'$\alpha={},\beta={},\gamma={}$'.format(alpha,beta,gamma))
Q = plt.quiver(diffop, bond, delta_o, delta_w, units='width')
#qk = plt.quiverkey(Q, 0.9, 0.9, 2, r'$2 \frac{m}{s}$', labelpos='E',
#                   coordinates='figure')

plt.xlabel('Difference in opinions')
plt.ylabel('Bond strength')
plt.show()
```
