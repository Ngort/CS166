```python
hash_dictionary = pd.DataFrame({'Name' : minerva_names,
'Hash' : [hashlib.sha224(bytes(name, 'utf-8')).hexdigest()[:5] for name in minerva_names]})

def unhash(hash):
    try:
        name = hash_dictionary.query('Hash=="{}"'.format(hash)).Name.values[0]
        return name
    except:
        return hash
```
